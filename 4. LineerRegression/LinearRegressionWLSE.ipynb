```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1) Veri yükleme
data = pd.read_csv('C:\\Users\\ASUS\\Downloads\\rounded_hours_student_scores.csv')
X = data[['Hours']].values      # (61,1)
Y = data['Scores'].values       # (61,)

# 2) Ortalama ve parametre hesaplama
x_mean = X.mean()
y_mean = Y.mean()

num = np.sum((X - x_mean) * (Y - y_mean).reshape(-1,1))
den = np.sum((X - x_mean)**2)

m = num / den
c = y_mean - m * x_mean

print(f"Eğim (m): {m:.4f}")
print(f"Kesim (c): {c:.4f}")

# 3) Tahmin ve görselleştirme
Y_pred = m * X + c

plt.scatter(X, Y, label='Gerçek Veriler')
plt.plot(X, Y_pred, color='red', label='En Küçük Kareler Doğrusu')
plt.xlabel('Çalışma Saati (Hours)')
plt.ylabel('Puan (Scores)')
plt.title('Manual Least Squares – Student Scores')
plt.legend()
plt.show()

# 4) Hata (MSE)
mse = np.mean((Y - Y_pred.flatten())**2)
print(f"MSE (WLS): {mse:.3f}")
```
