```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# 1) Veri yÃ¼kleme
data = pd.read_csv('C:\\Users\\ASUS\\Downloads\\rounded_hours_student_scores.csv')
X = data[['Hours']].values    # (61,1)
Y = data['Scores'].values     # (61,)

# 2) scikit-learn ile Linear Regression
model = LinearRegression()
model.fit(X, Y)

# 3) Tahmin ve metrikler
Y_pred = model.predict(X)
m = model.coef_[0]
c = model.intercept_
mse = mean_squared_error(Y, Y_pred)

print(f"[sklearn] EÄŸim (m): {m:.4f}")
print(f"[sklearn] Kesim (c): {c:.4f}")
print(f"[sklearn] MSE: {mse:.3f}")

# 4) Grafikte gÃ¶sterim
plt.scatter(X, Y, label='GerÃ§ek Veriler', edgecolor='k')
plt.plot(X, Y_pred, color='green', label='Sklearn Regression')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title('Student Scores â€“ scikit-learn LinearRegression')
plt.legend()
plt.show()
mse_skl= mse
```
